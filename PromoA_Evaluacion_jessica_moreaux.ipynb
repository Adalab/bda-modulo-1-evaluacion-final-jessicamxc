```python
import re

class TiendaOnline:
    
    def __init__(self):
        self.inventario = []
        self.clientes = {}
        self.ventas_totales = 0
    
    def agregar_productos(self, nombre, precio, cantidad):

        #si el inventario está vacío, el return es opcional
        if not self.inventario:
            #print("inventario vacío", nombre)
            self.inventario.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})
            return
     
        #si el inventario tiene productos
        for producto in self.inventario:
            if nombre == producto["nombre"]:
                #print("update", nombre)
                producto["cantidad"] = cantidad
                return
            
        #si no ha encontrado el producto en la lista de diccionarios, agrégalo
        #print("producto nuevo", nombre)
        self.inventario.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})

        
    def ver_inventario(self):
        for producto in self.inventario:
            print("Nombre:", producto["nombre"] ,"Precio: $", producto["precio"], "Cantidad: ", producto["cantidad"])

    def buscar_producto(self, nombre):
        #print("mi inventario actual", self.inventario)
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                print("Nombre:", producto["nombre"] ,"Precio: $", producto["precio"], "Cantidad: ", producto["cantidad"])

    def actualizar_stock(self, nombre, cantidad): 
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                producto["cantidad"] = cantidad

    def eliminar_producto(self, nombre):
        coincidencia= False
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                self.inventario.remove(producto)
                coincidencia= True
                print("Producto borrado")

        if not coincidencia:
            print("El producto no ha sido encontrado.")


    def calcular_valor_inventario(self):
        valor_inventario= 0
        # Hay que poner el contador fuera del bucle porque sino se inicializa a cero en el bucle.
        for producto in self.inventario:
            precio_total_productos= producto["precio"]
            cantidad_total_productos= producto["cantidad"]
            valor_inventario += precio_total_productos*cantidad_total_productos
        print("El valor del inventario es:", valor_inventario, "$")
        #el print tiene que estar fuera del for porque sino imprime el resultado en dos lineas y no por el total

    def  buscar_producto_regex(self, patron):
        resultado=[]
        for producto in self.inventario:
            if re.search(patron, producto["nombre"]):
                # es refindall la expresion mas correcta? o search
                resultado.append(producto)
        print(resultado)

    def realizar_compra(self):
        carrito=[]
        precio_total_inicial=0
        respuesta= input("¿Quieres comprar?")
        while respuesta == "si":
            print("Catalogo disponible:", self.inventario)
            compra_cliente= input("¿Que desea comprar, ingrese el nombre del producto deseado")
            for producto in self.inventario:
                if compra_cliente == producto["nombre"]:
                    producto["cantidad"]-=1
                    carrito.append(producto["nombre"])
                    precio_total= producto["precio"]
            respuesta= input("¿Quieres comprar?") 
            precio_total_inicial+= precio_total
        print("Resumen de su compra:", carrito, "precio total", precio_total_inicial ,"$")
            
    def procesar_pago(self):
        try:
            precio_compra= input("Ingrese el total de la compra")
            pago_cliente= input("Ingrese la cantidad total aportada")
            cambio= int(pago_cliente) - int(precio_compra)
            if cambio>0:
                print("Compra realizada con exito, cantidad a devolver al cliente", cambio, "$")
            elif cambio<0:
                print("Monto insuficiente, pendiente de cobro", cambio, "$")
            else:
                print("Compra realizada con exito")
        except:
            print("Ha ocurrido un error durante el proceso de pago. Intentelo mas tarde")

    def agregar_cliente(self, nombre, email):
        self.clientes[nombre]= {"email": email}

    def ver_clientes(self):
            for nombre, info_cliente in self.clientes.items():
                print("Nombre:", nombre, ", Correo:", info_cliente["email"])

    def registrar_compra(self, nombre_cliente, carrito):
        if nombre_cliente in self.clientes:
            print("El cliente está registrado")
            total_compra = sum(articulo['precio'] * articulo['cantidad'] for articulo in carrito.values())
            if 'registro_compras' not in self.clientes[nombre_cliente]:
                self.clientes[nombre_cliente]['registro_compras'] = []
            self.clientes[nombre_cliente]["registro_compras"].append({'productos': carrito, 'total': total_compra})
            self.ventas_totales += total_compra
        
        else:
            print("No se puede realizar la accion. Cliente no registrado")
    
    def ver_compras_cliente(self, nombre_cliente):
        if nombre_cliente in self.clientes:
            historial_cliente=self.clientes[nombre_cliente].get("registro_compras")
            print(historial_cliente)
            if not historial_cliente:
                print("Este cliente no tiene historial de compras")

        else:
            print("Cliente no se encuentra en nuestra base de datos")

    def calcular_ventas_totales(self):
        print("Las ventas totales son:", self.ventas_totales, "$")

                
```

```python
tienda_ropa= TiendaOnline()
tienda_ropa.agregar_cliente("Maria", "maria@gmail.com")
tienda_ropa.agregar_cliente("Roman", "romanperez@gmail.com")
tienda_ropa.ver_clientes()
print("------")
carrito1= {"Camisa": {"precio": 20,"cantidad": 3}}
carrito2= {"Pantalon": {"precio": 10, "cantidad": 4}}
tienda_ropa.registrar_compra("Maria", carrito1)
tienda_ropa.registrar_compra("Maria", carrito2)
print("------")
tienda_ropa.ver_compras_cliente("Maria")
print("--------")
tienda_ropa.calcular_ventas_totales()

p2= tienda_ropa.agregar_productos("pantalon", 10, 5)
p1= tienda_ropa.agregar_productos('camisa', 20, 40)
tienda_ropa.eliminar_producto("camisa")
```
